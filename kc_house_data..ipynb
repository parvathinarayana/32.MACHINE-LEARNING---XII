```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits
%matplotlib inline
```

```python
data=pd.read_csv('kc_house_data.csv')
data.head()
```

```python
# data understanding
data.info()
```

```python
# Create a correlation matrix
data.corr()
```

```python
cols_to_keep = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'sqft_living15', 'sqft_lot15']
data = data[cols_to_keep]

data.head()
```

```python
# Convert sqft_basement to binary
data['sqft_basement'] = data['sqft_basement'].map(lambda x: 1 if x>0 else x)
data.head()
```

```python
# Create a new feature comparing sqft_living to those nearby

data['sqft_living_ratio'] = data['sqft_living'].div(data['sqft_living15'], axis='index')
```

```python
from patsy import dmatrices
import statsmodels.api as sm

# fit the regression to eliminate values that are not statistically significant
y, X = dmatrices('price ~ bedrooms + bathrooms + sqft_living + sqft_lot + floors + C(waterfront) + C(view) + C(condition) + C(grade) + C(sqft_basement) + C(yr_built) + C(yr_renovated) + C(zipcode) + sqft_living15 + sqft_living_ratio', data, return_type = 'dataframe')

X.head()
```


## Modeling
We can now create our model after successfully preprocessing the data and use model metrics to determine how well the model can predict house prices.

```python
# Split the data into 70% Training Data and 30% Testing Data with seed(0)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Linear Regression
reg = LinearRegression()
regmodel = reg.fit(X_train, y_train)

coef = reg.coef_

y_predtrain = regmodel.predict(X_train)

y_predtest = regmodel.predict(X_test)
```

```python
RMSEtrain = mean_squared_error(y_train, y_predtrain)**0.5
RMSEtest = mean_squared_error(y_test, y_predtest)**0.5
r2_train = r2_score(y_train, y_predtrain)
r2_test = r2_score(y_test, y_predtest)
print ("RMSE for the training data is:", RMSEtrain)
print ("RMSE for the testing data is:", RMSEtest)
print ("R^2 value for training data is:", r2_train)
print ("R^2 value for testing data is:", r2_test)
```
